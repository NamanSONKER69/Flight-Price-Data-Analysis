# Flight Price Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_excel('Data_Train.xlsx')
df_train.head()
```

```python
df_train.info()
```

```python
df_test=pd.read_excel('Test_set.xlsx')
df_test.head()
```

```python
df_test.info()
```

```python
df_final=pd.merge(df_train,df_test,how='left')
df_final.head()
```

```python
df_final.info()
```

```python
df_final.isnull().sum()
```

```python
#Feature Engineering
# convert date in day, month, year
```

```python
df_final['Day']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year']=df_final['Date_of_Journey'].str.split('/').str[2]
df_final.head()
```

```python
df_final.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df_final.head()
```

```python
#ARRIVAL_TIME 
REMOVING DATE FROM TIME
```

```python
df_final['Arrival_Time']=df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final['Arrival_Hour']=df_final['Arrival_Time'].str.split(':').str[0]
```

```python
df_final['Arrival_Min']=df_final['Arrival_Time'].str.split(':').str[1]
```

```python
df_final.head()
```

```python
df_final['Dep_Hour']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dep_Min']=df_final['Dep_Time'].str.split(':').str[1]
```

```python
df_final.head()
```

```python
df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1,'NaN':1})
```

```python
df_final.head()
```

```python
df_final.drop(['Route'],axis=1,inplace=True)
df_final.drop(['Dep_Time'],axis=1,inplace=True)
df_final.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
df_final[df_final['Total_Stops'].isnull()]
```

```python
df_final.drop(['Duration'],axis=1,inplace=True)
df_final.drop(['Duration_Hour'],axis=1,inplace=True)
df_final.drop(['Duration_min'],axis=1,inplace=True)
```

```python
df_final.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
```
